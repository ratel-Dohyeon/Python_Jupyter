### 기초 그래프 그리기 - 히스토그램, 산점도, 박스그래프

```python
import seaborn as sns # 임포트 후, sns로 별명 줌 
tips = sns.load_dataset('tips')
print(tips.head())
print(type(tips))
# [지불금액,    팁,   성별, 흡연,  요일,  시간,  인원] 아래 컬럼 이름
```

```python
# fig(기본틀), 그래프  격자(axex1)
%matplotlib inline
import matplotlib.pyplot as plt

fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)
```

## 1. 히스토그램은 df의 열 데이터 분포와 빈도를 살펴보는 용도로 자주 사용
변수를 하나만 사용해서 그린 '일변량 그래프'

```python
axes1.hist(tips['total_bill'], bins=10) #bins=10 : x축의 간격을 10으로 조정
axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Freqiency')
axes1.set_ylabel('Total Bill')

fig
```

## 2. 산점도 그래프
변수 2개 사용한 '이변량 그래프' - total_bill 열에 따른 tip열의 분포를 나타낸 산점도 그래프

```python
# scatter 메서드에 total_bill, tips 열을 전달
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(tips['total_bill'], tips['tip'])
axes1.set_title('scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

## 3. 박스 그래프
* 이산형 병수 - Female, Male처럼 명확하게 구분되는 값을 의미
* 연속형 변수 - Tip와 같이 명확하게 셀 수 없는 변위의 값을 의미 

```python
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)
axes1.boxplot(
    [tips[tips['sex'] == 'Female']['tip'],
     tips[tips['sex'] == 'Male']['tip']],
    labels=['Fmale','Male'])

axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

## 다변량 데이터로 다변량 그래프 그리기 - 산점도 그래프
* 다변량 - 3개 이상의 변수를 사용

```python
# 성별을 새 변수로 추가(문자열은 산점도 그래프의 색상을 지정하는 값을 사용할 수 없다.)
def recode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1
```

```python
# apply 메서드는 사용자가 작성한 함수를 한번에 df의 각 행과 열에 적용하여 실핸 메서드
tips['sex_color'] = tips['sex'].apply(recode_sex)
```

```python
tips
```

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size'] *13, #size
    c=tips['sex_color'], #color
    alpha=0.7) #투명도

axes1.set_title('total_bill vs Tip Colored by sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

## 다변량 그래프 그리기
seaborn 라이브러리는 matlpotlib 라이브러리를 기반으로 만든 라이브러리

## seaborn 라이브러리로 히스토그램
* seaborn 라이브러리로 히스토그램을 그리려면 subplots, distplot 메서드를 사용
* distplot 메서드 : 기본 들 만듬 / total_bill 메서드 : 열데이터 전달

```python
ax = plt.subplots() #바탕
ax = sns.distplot(tips['total_bill']) #틀을 만들고 total_bill 전달  
ax.set_title('Total Bill Histogram With Density Plot')
```

```python
# 밀집도(kde-정규화시켜 넓이가 1이 되도록 그린 그래프) 제외

ax = plt.subplots() #바탕
ax = sns.distplot(tips['total_bill'], kde=False) #틀을 만들고 total_bill 전달  
ax.set_title('Total Bill Histogram')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Frequency')
```

```python
# 밀집도(kde-정규화시켜 넓이가 1이 되도록 그린 그래프) 제외

ax = plt.subplots() #바탕
ax = sns.distplot(tips['total_bill'], hist=False) #틀을 만들고 total_bill 전달  
ax.set_title('Total Bill Density')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Unit Probability')
```

```python
#rug인자 추가 - 그래프의 축에 동일한 길이의 직성을 붙여 데이터의 밀집 정도를 표현한 그래프 

hist_den_rug, ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug=True)
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')
```

## count 그래프 - 이산값을 나타낸 그래프

```python
# 이산값 - 명확하게 구분되는 값
ax = plt.subplots() #바탕
ax = sns.countplot('day', data = tips) #countplot에 있는 day 라는 것을 불러서 data로 tips 를 넣어라
ax.set_title('Count of days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```

## 다양한 종류의 이변량 그래프 그리기

### 1. seaborn 라이브러리로 산점도 그래프 그리기
* regplot 메서드를 사용하면 산점도 그래프와 회귀선을 그릴 수 있다.

```python
ax = plt.subplots() #바탕
ax = sns.regplot(x='total_bill', y='tip', data = tips) # 계속 표만 바꿔주고 있음 
ax.set_title('Scotterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax = plt.subplots() #바탕
ax = sns.regplot(x='total_bill', y='tip', data = tips, fit_reg=False) # fit_reg=False <- 회귀선 제거
ax.set_title('Scotterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
joint = sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
joint.set_axis_labels(xlabel = 'Total_Bill', ylabel = 'Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize = 10, y=1.03)
```

```python
#hexbin: 육각그래프 - 데이터를 쌓아 표현 
hexbin = sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = "hex")
hexbin.set_axis_labels(xlabel = 'Total_Bill', ylabel = 'Tip')
hexbin.fig.suptitle('Joint Plot of Total Bill and Tip', 
                    fontsize = 10, y=1.03)
```

## 4. 이차원 밀집도 그리기

```python
#열데이터 'total_bill', 'tip'
ax = plt.subplots() #바탕
ax = sns.kdeplot(data = tips['total_bill'],
                 data2 = tips['tip'],
                 shade=True) #음영효과, shade=False  
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot) # 대각선 기준 왼쪽(산점도 그래프)
pair_grid = pair_grid.map_lower(sns.kdeplot) # 대각선 기준 아래쪽(2차원 밀집도)
pair_grid = pair_grid.map_diag(sns.distplot, rug=True) # 대각선 중심(히스토그램, 밀집도)
plt.show()
```

```python
fig = plt.figure() 
seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks'] 
for idx, style in enumerate(seaborn_styles):
   # print(style)
    plot_position = idx + 1
   # print(plot_position)
    with sns.axes_style(style):
        ax = fig.add_subplot(2, 3, plot_position)
        violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
        violin.set_title(style) 
        
fig.tight_layout()
```

```python
import matplotlib.pyplot as plt

labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0,0.1,0.1,0) # 중심에서 벗어사는 정도 0.1 만큼 

fig, ax1 = plt.subplots()
ax1.pie(sizes, explode = explode, labels=labels, autopct = '%1.1f%%',
       shadow = True, startangle=90)
ax1.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0,0.1,0.1,0) # 중심에서 벗어사는 정도 0.1 만큼 \
colors = ['hotpink','green','yellow','blue']

fig, ax1 = plt.subplots()
ax1.pie(sizes, explode = explode, labels=labels, autopct = '%1.1f%%',
       shadow = True, colors =colors, startangle=90)
ax1.axis('equal')
plt.show()
```
